1. Создайте класс Length (Длина), имеющий свойства

• value (значение),

• unit (единица измерения).

При изменении единицы измерения значение должно соответственно меняться. Например, при переходе от сантиметров к метрам значение должно уменьшаться в 100 раз. Допустимые значения свойства unit: ‘см’, ‘м’, ‘км’. Организуйте эту проверку. Продемонстрируйте работу с классом.

```python
# Решение 1.
class Length:
    def __init__(self, value, unit):
        self.value = value
        self.unit = unit

    @property
    def value(self):
        return self._value

    @value.setter
    def value(self, value):
        self._value = value

    @property
    def unit(self):
        return self._unit

    @unit.setter
    def unit(self, unit):
        if unit not in ['см', 'м', 'км']:
            raise ValueError("Недопустимая единица измерения. Допустимые значения: 'см', 'м', 'км'.")
        self._unit = unit
        if unit == 'см':
            self._value = self._value
        elif unit == 'м':
            self._value = self._value / 100
        elif unit == 'км':
            self._value = self._value / 100000

# Демонстрация использования класса
length_in_cm = Length(200, 'см')
print(f"Длина в сантиметрах: {length_in_cm.value} {length_in_cm.unit}")

length_in_m = Length(2, 'м')
print(f"Длина в метрах: {length_in_m.value} {length_in_m.unit}")

length_in_km = Length(2, 'км')
print(f"Длина в километрах: {length_in_km.value} {length_in_km.unit}")

# Попытка использования неверной единицы измерения
try:
    length_in_invalid_unit = Length(2, 'мм')
except ValueError as e:
    print(e)
```

2. Создайте класс Ведомость, имеющий
атрибут класса:

• список_дисциплин (значением является список с названиями дисциплин);

свойства:

• дисциплина (при задании значения проверять наличие дисциплины в атрибуте список_дисциплин),

• группа;

методы:

• put – добавляет в ведомость информацию об оценке студента (фамилия, оценка – параметры метода). Для хранения данных внутри класса используйте словарь, в котором ключом является фамилия студента. Возможные оценки – «отлично», «хорошо», «удовл.», «неудовл.», «н/я»;

• get – возвращает оценку, полученную студентом (фамилия студента – параметр метода);

• change – изменяет оценку, полученную студентом (фамилия студента и новая оценка – параметры метода);

• del – удаляет информацию о студенте из ведомости (фамилия студента – параметр метода);

• result – возвращает кортеж из 5 чисел (количество каждого вида оценок в ведомости);

• __init__ – конструктор;

• __str__ – возвращает строку, содержащую заголовки (название экзамена, группа) и результаты экзамена в виде таблицы;

• count – возвращает количество студентов в ведомости;

• names – возвращает список фамилий, имеющихся в ведомости.

Продемонстрируйте работу с классом. 


```python
# Решение 2.
class Ведомость:
    список_дисциплин = ["Математика", "Физика", "Информатика", "Английский язык", "Химия"]

    def __init__(self, дисциплина, группа):
        self.дисциплина = дисциплина
        self.группа = группа
        self.оценки = {}

    @property
    def дисциплина(self):
        return self._дисциплина

    @дисциплина.setter
    def дисциплина(self, value):
        if value in self.список_дисциплин:
            self._дисциплина = value
        else:
            raise ValueError(f"Дисциплина '{value}' не входит в список_дисциплин: {self.список_дисциплин}")

    def put(self, фамилия, оценка):
        if оценка in ["отлично", "хорошо", "удовл.", "неудовл.", "н/я"]:
            self.оценки[фамилия] = оценка
        else:
            raise ValueError("Недопустимая оценка. Можно использовать: 'отлично', 'хорошо', 'удовл.', 'неудовл.', 'н/я'.")

    def get(self, фамилия):
        if фамилия in self.оценки:
            return self.оценки[фамилия]
        else:
            raise KeyError(f"Студент с фамилией '{фамилия}' не найден в ведомости.")

    def change(self, фамилия, новая_оценка):
        if фамилия in self.оценки:
            if новая_оценка in ["отлично", "хорошо", "удовл.", "неудовл.", "н/я"]:
                self.оценки[фамилия] = новая_оценка
            else:
                raise ValueError("Недопустимая оценка. Можно использовать: 'отлично', 'хорошо', 'удовл.', 'неудовл.', 'н/я'.")
        else:
            raise KeyError(f"Студент с фамилией '{фамилия}' не найден в ведомости.")

    def delete(self, фамилия):
        if фамилия in self.оценки:
            del self.оценки[фамилия]
        else:
            raise KeyError(f"Студент с фамилией '{фамилия}' не найден в ведомости.")

    def result(self):
        отлично = 0
        хорошо = 0
        удовл = 0
        неудовл = 0
        н_я = 0

        for оценка in self.оценки.values():
            if оценка == "отлично":
                отлично += 1
            elif оценка == "хорошо":
                хорошо += 1
            elif оценка == "удовл.":
                удовл += 1
            elif оценка == "неудовл.":
                неудовл += 1
            elif оценка == "н/я":
                н_я += 1

        return (отлично, хорошо, удовл, неудовл, н_я)

    def __str__(self):
        result = f"Экзамен по дисциплине: {self.дисциплина}\nГруппа: {self.группа}\n\n"
        result += "Фамилия | Оценка\n"
        result += "-" * 30 + "\n"

        for фамилия, оценка in self.оценки.items():
            result += f"{фамилия:20} | {оценка}\n"

        return result

    def count(self):
        return len(self.оценки)

    def names(self):
        return list(self.оценки.keys())
```

```python
# Создание ведомости
ведомость = Ведомость("Математика", "БПМ-21")

# Добавление оценок
ведомость.put("Иванов", "отлично")
ведомость.put("Петров", "хорошо")
ведомость.put("Сидоров", "удовл.")
ведомость.put("Кузнецов", "неудовл.")
ведомость.put("Смирнов", "н/я")

# Получение оценки
print(ведомость.get("Петров"))  # Вывод: хорошо

# Изменение оценки
ведомость.change("Иванов", "хорошо")

# Удаление студента
ведомость.delete("Кузнецов")

# Получение результатов
print(ведомость.result())  # Вывод: (1, 2, 1, 0, 1)

# Вывод ведомости
print(ведомость)
# Вывод:
# Экзамен по дисциплине: Математика
# Группа: БПМ-21
#
# Фамилия | Оценка
# ------------------------------
# Иванов         | хорошо
# Петров        | хорошо
# Сидоров       | удовл.
# Смирнов       | н/я

# Получение количества студентов
print(ведомость.count())  # Вывод: 4

# Получение списка фамилий
print(ведомость.names())  # Вывод: ['Иванов', 'Петров', 'Сидоров', 'Смирнов']
```

3. Создайте класс Заказ(Order), у которого есть свойства код_товара(code), цена(price), количество(count) и методы __init__ и __str__.

Создайте 2 класса-потомка: Опт(Opt) и Розница(Retail).
В этих классах создайте методы __init__, __str__.и сумма_заказа(summa), позволяющий узнать стоимость заказа.
Для опта стоимость единицы товара составляет 95% от цены, а при покупке более 500 штук – 90% цены.
В розницу стоимость единицы товара составляет 100% цены. Стоимость заказа равна произведению цены на количество. 

Продемонстрируйте работу с классами, создав необходимые объекты и обратившись к их свойствам и методам


```python
# Решение 3.
class Order:
    def __init__(self, code, price, count):
        self.code = code
        self.price = price
        self.count = count

    def __str__(self):
        return f"Код товара: {self.code}, Цена: {self.price}, Количество: {self.count}"


class Opt(Order):
    def __init__(self, code, price, count):
        super().__init__(code, price, count)

    def __str__(self):
        return super().__str__()

    def summa(self):
        if self.count > 500:
            price_per_unit = self.price * 0.9
        else:
            price_per_unit = self.price * 0.95
        return price_per_unit * self.count


class Retail(Order):
    def __init__(self, code, price, count):
        super().__init__(code, price, count)

    def __str__(self):
        return super().__str__()

    def summa(self):
        return self.price * self.count

opt_order = Opt(1234, 100, 600)
print(opt_order)
print(f"Сумма заказа: {opt_order.summa()}")

retail_order = Retail(5678, 100, 10)
print(retail_order)
print(f"Сумма заказа: {retail_order.summa()}")
```

4. Написать функцию my_func_1(lst), которая переворачивает список вложенных списков с произвольным уровнем вложенности.
Любой элемент, не являющийся списком, не должен быть изменен.

Примеры, демонстрирующие ожидаемое поведение функции:

assert my_func_1(42) == 42

assert my_func_1([]) == []

assert my_func_1([1, 2]) == [2, 1]

assert my_func_1((1, 2)) == (1, 2)

assert my_func_1([[1, 2], [3, [4, 5]], 6]) == [6, [[5, 4], 3], [2, 1]]

assert my_func_1(['123', '456', '789', {10: [11, 12], 13: 14}, (15, 16)]) == [(15, 16), {10: [11, 12], 13: 14}, '789', '456', '123']

assert my_func_1([[[[[1, 2]]]]]) == [[[[[2, 1]]]]]

```python
# Решение 4.
def my_func_1(lst):
    if isinstance(lst, list):
        return [my_func_1(x) for x in reversed(lst)]
    else:
        return lst
```

```python
assert my_func_1(42) == 42
assert my_func_1([]) == []
assert my_func_1([1, 2]) == [2, 1]
assert my_func_1((1, 2)) == (1, 2)
assert my_func_1([[1, 2], [3, [4, 5]], 6]) == [6, [[5, 4], 3], [2, 1]]
assert my_func_1(['123', '456', '789', {10: [11, 12], 13: 14}, (15, 16)]) == [(15, 16), {10: [11, 12], 13: 14}, '789', '456', '123']
assert my_func_1([[[[[1, 2]]]]]) == [[[[[2, 1]]]]]
```

5.При помощи механизма map/filter/reduce возвести в квадрат числа от 1 до 100, и рассчитать их сумму, не включая в сумму числа, кратные 9. Написать реализацию в одну строку.

```python
# Решение 5.
sum(map(lambda x: x**2, filter(lambda x: x % 9 != 0, range(1, 101))))
```
